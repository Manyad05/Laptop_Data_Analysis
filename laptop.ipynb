```python
# Replace "Missing" with an empty string in the entire DataFrame
df.replace("Missing", "", inplace=True)

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Plot 1: Brand Distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='brand', data=df, order=df['brand'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.title('Brand Distribution')
plt.show()
```

```python
# Plot 2: Average Star Rating by Brand
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='star_rating', data=df, ci=None, order=avg_star_rating_by_brand.index)
plt.xticks(rotation=45, ha='right')
plt.title('Average Star Rating by Brand')
plt.show()
```

```python
# Plot 3: Operating System Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='os', data=df, order=df['os'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.title('Operating System Distribution')
plt.show()
```

```python
# Plot 4: Relationship between OS and Star Rating
plt.figure(figsize=(8, 5))
sns.barplot(x='os', y='star_rating', data=df, ci=None, order=os_rating_relation.index)
plt.xticks(rotation=45, ha='right')
plt.title('Relationship between Operating System and Star Rating')
plt.show()
```

```python
# Plot 5: Relationship between Processor Generation and Star Rating
plt.figure(figsize=(10, 5))
sns.barplot(x='processor_gnrtn', y='star_rating', data=df, ci=None, order=processor_rating_relation.index)
plt.xticks(rotation=45, ha='right')
plt.title('Relationship between Processor Generation and Star Rating')
plt.show()
```

```python
# Plot 6: Distribution of SSD and HDD
plt.figure(figsize=(8, 5))
sns.countplot(x='ssd', data=df)
plt.title('Distribution of SSD')
plt.show()
```

```python
# Plot 2: Pairplot for Numerical Columns
numerical_cols = ['ram_gb', 'weight', 'display_size', 'latest_price', 'old_price', 'star_rating', 'ratings', 'reviews']
sns.pairplot(df[numerical_cols])
plt.suptitle('Pairplot for Numerical Columns', y=1.02)
plt.show()
```

```python
# Plot 3: Heatmap for Correlation Matrix
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.set(style="whitegrid")

# Pie Chart 1: Distribution of Operating Systems
plt.figure(figsize=(8, 8))
os_distribution = df['os'].value_counts()
plt.pie(os_distribution, labels=os_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Operating Systems')
plt.show()
```

```python
# Pie Chart 2: Distribution of Touchscreen
plt.figure(figsize=(6, 6))
touchscreen_distribution = df['Touchscreen'].value_counts()
plt.pie(touchscreen_distribution, labels=touchscreen_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Touchscreen')
plt.show()
```

```python
# Donut Chart: Distribution of Warranty Durations
plt.figure(figsize=(8, 8))
warranty_distribution = df['warranty'].value_counts()
plt.pie(warranty_distribution, labels=warranty_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.gca().add_artist(plt.Circle((0,0),0.70,fc='white'))
plt.title('Distribution of Warranty Durations')
plt.show()
```

```python
# Horizontal Bar Plot: Distribution of Brands
plt.figure(figsize=(12, 8))
brand_distribution = df['brand'].value_counts()
sns.barplot(x=brand_distribution, y=brand_distribution.index, palette="viridis")
plt.title('Distribution of Brands')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.show()
```
